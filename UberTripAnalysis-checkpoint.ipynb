## Uber Trip Analysis

##### Uber Trip Analysis is a Analysis how many uber trips we can draw many patterns like which day has the highest and the lowest trips or the busiest hour for uber and many other patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## reading csv
data = pd.read_csv("C:/Users/Nikita/Downloads/uber-raw-data-sep14.csv/uber-raw-data-sep14.csv")
```

```python
data.head()
```

```python
data["Date/Time"] = data["Date/Time"].map(pd.to_datetime)
```

```python
data["Date/Time"]
```

```python
data["Day"] = data["Date/Time"].apply(lambda x: x.day)
```

```python
data["Day"] 
```

```python
data["Weekday"] = data["Date/Time"].apply(lambda x: x.weekday())
```

```python
data["Weekday"]
```

```python
data["Hour"] = data["Date/Time"].apply(lambda x: x.hour)
```

```python
data["Hour"] 
```

```python
data.head()
```

```python
## Plotting graphs using seaborn
sns.set(rc={'figure.figsize':(12, 10)})
sns.distplot(data["Day"])
```

```python
sns.distplot(data["Hour"])
```

```python
sns.distplot(data["Weekday"])
```

```python
df = data.groupby(["Weekday", "Hour"]).apply(lambda x: len(x))
df = df.unstack()
```

```python
sns.heatmap(df, annot=False)
```

```python
data.plot(kind='scatter',x='Lon',y='Lat',alpha = 0.4, s=data['Day'], label ='Uber Trips',figsize = (12,8),cmap = plt.get_cmap('jet'))
```
